## Imports

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd           
import seaborn as sns
```

## Read-in Data

```python
ames = pd.read_csv('../data/Ames_coefficients.csv', keep_default_na=False, na_values=[''])  
train = pd.read_csv('../data/train_clean.csv', keep_default_na=False, na_values=[''])  
```

```python
ames_n = ames.iloc[10:29, :]
```

```python
ames_n = ames_n.sort_values('coef')
```

```python
ames_n['positive'] = ames_n['coef'] > 0
```

# Visuals

## Notable Ames Neighborhoods

```python
# Size
plt.figure(figsize = (8, 8))

# Bar Chart - Horizontal 
plt.barh(ames_n['Neighborhood'], ames_n['coef'], color=ames_n.positive.map({True: 'blue', False: 'orange'}))


plt.title('Notable Ames Neighborhoods', fontsize=20)

plt.xlabel('$Price Difference from other neighborhoods', fontsize=14) # Add axis labels
plt.savefig('../images/Notable_Ames_Neighborhoods.png');
```

## Sale Price by SQ Ft

```python
x=train['SF']

plt.figure(figsize=(8, 8))
plt.scatter(train['SF'], train['SalePrice'])

m, b = np.polyfit(train['SF'],
            train['SalePrice'], 1)


#add linear regression line to scatterplot 
plt.plot(x, m*x+b, color = "#a7bbc7")

# Create a descriptive title
plt.title('Sale Price by SQ Ft of home', fontsize=20)

# Add axis labels
plt.ylabel('$Sale Price', fontsize=14)
plt.xlabel('Square Feet', fontsize=14);

# Save Fig
plt.savefig('../images/SquareFeet_price.png');
```

## Sale Price by Year Built

```python
x=train['Year Built']

plt.figure(figsize=(8, 8))
plt.scatter(train['Year Built'], train['SalePrice'])

m, b = np.polyfit(train['Year Built'],
            train['SalePrice'], 1)


#add linear regression line to scatterplot 
plt.plot(x, m*x+b, color = "#a7bbc7")

# Create a descriptive title
plt.title('Sale Price by Year Built', fontsize=20)

# Add axis labels
plt.ylabel('$Sale Price', fontsize=14)
plt.xlabel('Year Built', fontsize=14);

# Save Fig
plt.savefig('../images/Year Built_price.png');
```
